```python
#Read the titanic dataset into pandas data frame titanic
import pandas as pd
url= 'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
titanic.head(5)
```

```python
#1. Create a pie chart presenting the male/female proportion
import matplotlib.pyplot as plt
%matplotlib inline

labels = ['male','female']
sizes = titanic.sex.value_counts()
print(labels[0],sizes[0])
print(labels[1],sizes[1])
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.2f%%', shadow=True)
#ax1.axis('equal')
plt.show()
```

```python
#2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender
#Matplotlib plot

plt.figure(figsize=(12,10))
ax = titanic[titanic.sex=='female'].plot.scatter('age', 'fare', color ='blue', label='female')
titanic[titanic.sex=='male'].plot.scatter('age', 'fare', ax = ax, color ='red', label = 'male')
```

```python
#Seaborn plot

import seaborn as sns
plt.figure(figsize=(10,6))
sns.stripplot(x='age',y='fare',data=titanic,hue='sex',dodge=True)
```
